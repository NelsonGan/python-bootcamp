```python
my_nums = [1,2,3,4,5]
```

```python
for item in map(lambda num: num**2,my_nums):
    print(item)
```

```python
list(map(square,my_nums))
```

```python
def splicer(mystring):
    if len(mystring) % 2 == 0:
        return 'EVEN'
    else:
        return mystring[0]
```

```python
names = ['Andy','Eve','Sally']
```

```python
for name in map(splicer,names):
    print(name)
```

```python
def check_even(num):
    return num % 2 == 0
```

```python
mynums = [1,2,3,4,5,6]
```

```python
list(filter(check_even,mynums))
```

```python
for n in filter(check_even,mynums):
    print(n)
```
