```python
class Dog():
    
    def __init__(self,breed,name,spots):
        
        # Attributes
        # Take in argument (dogbreed) and assign it using self.attribute_name (self.bread)
        self.breed = breed
        self.name = name
        
        #Expect Boolean
        self.spots = spots
```

```python
my_dog = Dog(breed='Huskie',name='Blackie',spots=False)
```

```python
type(my_dog)
```

```python
my_dog.breed
```

```python
my_dog.name
```
